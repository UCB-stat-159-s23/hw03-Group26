```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
import xarray as xr

from pathlib import Path

# Small style adjustments for more readable plots
plt.style.use("seaborn-whitegrid")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["font.size"] = 14
```

```python
#Read in the C02 csv file and display it.

DATA_DIR = Path.home()/Path('shared/climate-data')

co2_file = DATA_DIR / "monthly_in_situ_co2_mlo_cleaned.csv"
co2data = pd.read_csv(co2_file, na_values=-99.99).dropna()
co2data
```

The "hue" argument in the sns.lineplot() is used to group variables together and helps depict the relationship between the x and y variables. In this case, we want to group the x and y axes (months and C02 levels, respectively) together and group them via years. 

```python
#Create plot that shows "monthly cycle for all three years in the dataset." Question 1.

sns.lineplot(data=co2data, x="month", y="c02", hue="year")
```
